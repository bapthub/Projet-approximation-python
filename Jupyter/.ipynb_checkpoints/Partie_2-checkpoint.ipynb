# 3. Méthode pour tracer le polynôme avec la matrice de Vandermone

```python
def vander_method(l_x,l_y):
    m = np.zeros((len(l_x),len(l_x)),dtype=int)
    for i in range(len(m)):
        for j in range(len(m[i])):
            if i == 0:
                m[i][j] = l_x[i]
            else:
                m[i][j] = l_x[i]**j
    inv_matrix_v = inv(m)
    coeff = inv_matrix_v @ l_y
    return coeff
```

# 4. Méthode pour tracer le polynôme avec l'interpolation de Lagrange

```python
def Lk(x, k, coordonnee):
    result = 1
    for i in range(len(coordonnee)):
        if (i != k):
            result *= (x - coordonnee[i][0]) / (coordonnee[k][0] - coordonnee[i][0])
    return result
```

```python
def polynomeLagrange(x, coordonnee):
    result = 0
    for k in range(len(coordonnee)):
        result += Lk(x, k, coordonnee) * coordonnee[k][1]
    return result
```

```python
def tracer_lagrange(f, start, end, coordonnee):
    X = np.arange(start, end + pas, pas)
    Y = [ eval(f, x) for x in X ]
    plt.title("Function " + str(f))
    plt.plot(X, Y)
    for cooX, cooY in coordonnee:
        plt.scatter(cooX, cooY, color="black", marker="x")
    plt.axhline(0, color = "black")
    plt.axvline(0, color = "black")
    plt.grid()
    plt.show()
```

# 5. Interpolation de Newton via la matrice triangulaire et tablea des différences divisées

```python
def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    # the first column is y
    coef[:,0] = y

    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
    return coef

def calculatate_prod(l_x,i,j):
    current_x = l_x[i]
    res = 1
    for k in range(j):
        res *= current_x - l_x[k]
    return res

def newton_method(l_x,l_y):
    m = np.zeros((len(l_x),len(l_x)),dtype=int)
    for i in range(len(m)):
        for j in range(len(m)):
            if j == 0:
                m[i][j] = 1
            else:
                m[i][j] = calculatate_prod(l_x,i,j)
    inv_matrix_v = inv(m)
    coeff = inv_matrix_v @ l_y
    return coeff
```

# 6. Tester les méthodes sur les exercices 5

```python
coordonnee1 = [
               (1, 1), 
               (2, 8), 
               (3, 27),
               (4, 64)
               ]
```

```python
x5 = np.array([1, 2, 3, 4])
y5 = np.array([1, 8, 27, 64])
```

##### Matrice Vandermonde

```python
coeff = vander_method(x5,y5)
# round at 3 digits
for i in range(len(coeff)):
        coeff[i] = round(coeff[i],3)
print(str(int(coeff[3])) + " x^3 " + " + "  + str(coeff[2]) + " x^2 " + str(coeff[1]) + " x " + " + " + str(coeff[0]))
#plot
lower_limit = 0
upper_limit = 8
num_pts = 100
poly_coefs = coeff[::-1]
x_line = np.linspace(lower_limit, upper_limit, num_pts)
y_new = np.polyval(poly_coefs, x_line)
plt.figure(figsize = (10, 6))
plt.plot(x5, y5, 'bo')
plt.plot(x_line, y_new, '-b')
plt.show()
```

##### Interpolation de Lagrange

```python
f1 = polynomeLagrange(x, coordonnee1)
print(f1)
tracer_lagrange(f1, 0, 5, coordonnee1)
```
